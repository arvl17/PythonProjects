```python
#Classification of Movies based on Ratings

def rating(rating):
    if rating >= 8.0:
        return 'Excellent'
    elif rating >= 6.0:
        return 'Good'
    else:
        return 'Average'
```

```python
df['Rating_Category'] = df['Rating'].apply(rating)
df.head()
```

```python
#Total action movies

len(df[df['Genre'].str.contains('Action', case = False)])
```

```python
#Unique values from Genre

list1 = []
for value in df ['Genre']:
    list1.append(value.split(','))

list1
```

```python
d = []
for item in list1:
    for item1 in item:
        d.append(item1)
        
d
```

```python
unique = []
for item in d:
    if item not in unique:
        unique.append(item)

unique
```
